```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t=np.linspace(0,.003,3000)
a=np.cos(2000*np.pi*t)
plt.plot(t,a)
plt.ylabel('Magnitude')
plt.xlabel('Time')
plt.title('Continuous Signal')
plt.show()
```

```python
n=np.linspace(0,9,10)
an=np.cos(2*np.pi*n/3)
plt.stem(an)
plt.ylabel('Magnitude')
plt.xlabel('Time Index')
plt.title('Discrete Signal')
plt.show()
```

```python
t=np.linspace(0,.004,3000)
b=np.sin(2000*np.pi*t)
plt.plot(t,b)
plt.ylabel('Magnitude')
plt.xlabel('Time')
plt.title('Continuous Signal')
plt.show()
```

```python
n=np.linspace(0,6,7)
bn=np.sin(2*np.pi*n/3)
plt.stem(bn)
plt.ylabel('Magnitude')
plt.xlabel('Time Index')
plt.title('Discrete Signal')
plt.show()
```

```python
t=np.linspace(0,.004,3000)
c=np.sin(2000*np.pi*t)/(np.pi*t)
c[0]=2000
plt.plot(t,c)
plt.ylabel('Magnitude')
plt.xlabel('Time')
plt.title('Continuous Signal')
plt.show()
```

```python
n=np.linspace(0,20,21)
cn=np.sin(2*np.pi*n/5)/(np.pi*n/5000)
plt.stem(cn)
plt.ylabel('Magnitude')
plt.xlabel('Time Index')
plt.title('Discrete Signal')
plt.show()
```
