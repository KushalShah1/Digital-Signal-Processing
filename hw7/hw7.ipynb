```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=np.linspace(0,31,32)
x=10*np.cos(np.pi*n/16.0)
y=[]
for i in x:
    if(i>8):
        y.append(8)
    elif(i<-8):
        y.append(-8)
    else:
        y.append(i)
```

```python
Xk=np.fft.fft(x)
Yk=np.fft.fft(y)

plt.stem(Xk)
plt.title("X[k]")
plt.show()
```

```python
plt.stem(Yk)
plt.title("Y[k]")
plt.show()
```

# The DFT of x[n] is very uniform with only spike. This is accurate since there is only one frequency in x[n]. However the DFT of y[n] contains some wiggles around the core peak. This indicates there are other frequencies at play but there isn't.
